# Chapter 2 - Cointegration and Arbitrage Trading on Dual-Class Shares from Brazil Market
#### Cointegration and mean reverse strategy applied to different shares classes from the same bank, Itaú Unibanco in Brazil.

### Introduction

Pair trading is a popular statistical arbitrage strategy, it consists in trading the spread (long/short position) of shares, which usually possess the same "economic" factors, when a dislocation between the two prices paths is observed. To set up such strategy we split the implementation in 2 parts: the selection part (which pairs combination provide the right conditions (stationary state)) and the implementation of the trading strategy with back-test so an analyze on cumulative P&L behavior is possible.

In this chapter we introduce the concept of stationarity of a linear combination of assets, leading to cointegration and unit root tests to determine if a pair of Brazilian bank stocks of different classes for the same bank (ITUB4 and ITUB3) are suitable for a mean-reverting trade strategy. On a second step a third factor (asset) is introduced on the strategy, the objective is to perform a multivariate cointegration through a Johansen Procedure and explore if it is possible to improve the back-test results with the additional asset on the portfolio. The same P&L analysis is implemented, and results compared for both cases.

*The implementation was made in Python 3.6.3 using Jupyter Notebook, explanation of the code is not in scope of this chapter, please refer to the Python Notebook BanksArbitrage.ipynb for a commented version of the code implementation


### Dual Class Shares in Brazil - A Background

In the Brazilian financial markets, a considerable number of companies have 2 classes of shares traded on the exchange: **Ordinary shares (ON)** and **preference shares (PN)**:

**Common Shares** - Gives its holder the right to vote in shareholders' meetings and participate in the economic profit of the company. It is represented by the number three at the end of the share abbreviation, in our example: ITUB3

**Preference Shares$^1$** - Grants its holder preference in the receipt of dividends and, in the case of liquidation, preference in capital reimbursement. It does not give the holder the right to vote in shareholders’ meetings. They are represented by the number four, five, six, seven or eight on the stock code, in our example: ITUB4

As described on *An Assessment of Dual-Class Shares in Brazil: Evidence from the Novo Mercado Reform* by Matt Orsagh *(https://blogs.cfainstitute.org/marketintegrity/2017/05/09/an-assessment-of-dual-class-shares-in-brazil-evidence-from-the-novo-mercado-reform/),* historically, Brazilian companies have had a dual-class structure with common shares (voting) and preferred shares (nonvoting). To compensate for no voting rights, preferred shares pay (in theory) higher dividends or have "tag-along" rights (takeout rights on a sale of control). This enables holders of voting shares to control companies by owning less of the company’s total equity, which is a deviation from the proportionality principle ("one-share one-vote").

The preference shares have been introduced in Brazilian markets around mid-1970 in an attempt to incentivize more companies to open its capital on the exchange without giving up control, the initial law of 1976 allowed companies to issue 2/3 of total shares available as preference shares. This instrument was widely used by companies throughout the 80s, 90s until beginning of years 2000 and it PN shares quickly became the most liquid class and most of the times the only option for big institutional investors.

By earlier 2000 with a more mature market a process to decrease the size of preference shares was initiated with a change on the 1976 law decreasing the PN proportion to 50/50 on the subsequent years new rules were introduced with the objective to extinguish the market of PNs especially the rules about governance levels which states that for a company to be included on the exchange highest level of governance they should not have any PN traded on the market. The companies that inherited this characteristic are exchanging PN shares for ON, but there are still a lot of companies that possess both classes, especially on the banking industry

On the example on this chapter our choice was to use the 2 classes from Itaú Unibanco (https://www.itau.com.br/relacoes-com-investidores/o-itau-unibanco/sobre-o-itau-unibanco), the bank is the result of the merger of Banco Itaú and Unibanco, which occurred on November 4, 2008 to form Itaú Unibanco Holding S.A, the largest financial conglomerate in the Southern Hemisphere and is the 10th largest bank in the world by market value. The bank its listed at the B3 in São Paulo and in NYSE in New York. It currently is the biggest Latin American bank by assets and market capitalization.

Although in theory the differences of rights on the share classes would justify the premium/discount that are observed in the market in practice on Itaú Unibanco case it does not happen since dividends and tag-along rights are the same for both classes, still the preference class have historically traded ober a premium in relation to the common type, the two factors that are left to explain the premium/discount relationship are value of shareholders rights and the value of liquidity of the shares.

As expected those shares hold a strong positive relationship (chart 2), our first objective is to understand if a linear combination between these 2 stocks produces a stationary time series that can be translated in a opportunity for a mean-reverting strategy. The second objective is to try to improve the strategy’s result by optimizing improving its construction methodology. Scatter plots (chart 2 and 3) of the 2 stocks suggest a strong positive correlation, but it is not a guarantee of stationarity between the time series, but if we take a linear combination of these series it can sometimes lead to a stationary series. In this case we say that the series are cointegrated. The mathematical definition: Let xt and yt be two stationary time series, with a, b as constants. If the resulting series of axt + byt is stationary, then we say that xt and yt are cointegrated.

```python
class DataFrameSizeError(Exception):
    pass


class CointegrationCADF(object):

    def __init__(self, timeserie):
        if timeserie.shape[1] <= 1:
            raise DataFrameSizeError('DataFrame must have at least 2 TimeSeries')
        timeserie = timeserie.dropna()
        self._names = list(timeserie.columns.values)
        self._timeserie = timeserie
        self._col_pairs = list(itertools.permutations(self._timeserie.columns, 2))

    def pair_cadf(self, regression_results):

        cadfresults = []
        for index, row in regression_results.iterrows():
            dep_variable = self._timeserie[row[0]]
            indep_variable = self._timeserie[row[1]]
            hedge_ratio = row['hedge_ratio']
            spread = (dep_variable - hedge_ratio*indep_variable).reshape(dep_variable.shape[0])
            cadf = stt.adfuller(spread)
            cadfresults.append(cadf[0])
        regression_results['cadf_stat'] = cadfresults
        regression_results['10pct_confidence'] = regression_results.cadf_stat < cadf[4]['10%']
        regression_results['5pct_confidence'] = regression_results.cadf_stat < cadf[4]['5%']
        regression_results['1pct_confidence'] = regression_results.cadf_stat < cadf[4]['1%']
        return regression_results

    def ols_regression(self, perform_cadf=None, return_positive=None) -> object:
        if perform_cadf is None:
            perform_cadf=False
            return_positive=False
        if not perform_cadf: return_positive == False
        models = []
        results = []
        for a, b in self._col_pairs:
            model = sm.OLS(self._timeserie[a], self._timeserie[b], missing='drop')
            result = model.fit()
            results.append([a, b, result.params[0]])
        results = pd.DataFrame(results)
        results.columns = (['dep_variable', 'inded_variable', 'hedge_ratio'])
        if perform_cadf:
            results = self.pair_cadf(results)
            if return_positive:
                results = results[results.iloc[:, 4]]
                results=results.reset_index()
                return results
            else:
                return results
        else:
            return results

def half_life_cadf(ts):
    
    #Implemented a halflife calculation based on regression results from the spread mean - spread
    #And delta the spread
    
    spread_x = np.mean(ts.sprd) - ts.sprd 
    spread_y = ts.sprd.shift(-1) - ts.sprd
    spread_df = pd.DataFrame({'x':spread_x,'y':spread_y})
    spread_df = spread_df.dropna()

    #Theta as regression beta between pair spread and difference between pair spread
    model_s = sm.OLS(spread_df['y'], spread_df['x'])
    model_s = model_s.fit() 
    theta=  model_s.params[0]
    hl = math.log(2)/theta
    return hl

def gen_strategy(sprd_port,lookback,z_entry,z_close):
    
    #Create a z-score of the spread
    sprd_port['moving_average'] = sprd_port.sprd.rolling(lookback).mean()
    sprd_port['moving_std_dev'] = sprd_port.sprd.rolling(lookback).std()
    sprd_port=sprd_port.dropna()
    sprd_port['z_score'] = (sprd_port.sprd- sprd_port.moving_average)/sprd_port.moving_std_dev
    
    #Set up entry long and short points and exit from positions
    sprd_port['long_entry'] = (sprd_port.z_score < -z_entry)*1.0
    sprd_port['short_entry'] = (sprd_port.z_score > z_entry)*1.0  
    sprd_port['long_exit'] = (sprd_port.z_score >= -z_close)*1.0 
    sprd_port['short_exit'] = (sprd_port.z_score <= z_close)*1.0
    
    #Consolidating Entrys and Exits to return final position
    sprd_port['positions_long'] = np.nan  
    sprd_port.positions_long.loc[sprd_port.long_entry==1]=1  
    sprd_port.positions_long.loc[sprd_port.long_exit==1]=0    
    sprd_port['positions_short'] = np.nan  
    sprd_port.positions_short.loc[sprd_port.short_entry==1]=-1
    sprd_port.positions_short.loc[sprd_port.short_exit==1]=-0
    sprd_port = sprd_port.fillna(method='ffill')  
    sprd_port['positions'] = sprd_port.positions_long + sprd_port.positions_short
    sprd_port=sprd_port.fillna(0)
    
    #Cleaning the Dataframe
    sprd_port=sprd_port.drop(['moving_average','moving_std_dev'], axis=1)
    return sprd_port

def gen_back_test(backtest,hedge_ratio,show_entire_df,initialcash,trans_cost):
    
    ###This is my implementation of a Back Test based on a NAV approach, this is almost entirally vectorised###
    
    
    #defining dependent and independet asset price, independent is multiplied by hedge ratio
    backtest['dep_price'] = backtest['ITUB4']
    backtest['indep_price'] = backtest['ITUB3']*hedge_ratio
    
    #defining short deltas this is used to make short adjustments to NAV
    backtest['dep_short_adj']= backtest.dep_price.shift(-1)-backtest.dep_price
    backtest['indep_short_adj']= backtest.indep_price.shift(-1)-backtest.indep_price
    #declaring new columns on the dataframe with value = 0
    backtest['entry_long_cash'] = 0.0
    backtest['entry_short_cash'] = 0.0
    backtest['exit_long_cash'] = 0.0
    backtest['exit_short_cash'] = 0.0
    backtest['initial_cash'] = 0.0
    backtest['end_cash'] = 0.0
    backtest['long_nav'] = 0.0
    backtest['short_nav'] = 0.0
    backtest['long_nav'] = 0.0
    #Cash effect on A New long position is dependent on flag to open a long and if we are not alredy long, effect is cah negative since we are purchasing the dependent stock
    backtest.entry_long_cash[(backtest.long_entry==1) & ((backtest.positions.shift(1)==0) | (backtest.positions.shift(1)==-1))]=-backtest.dep_price.shift(-1)
    #Cash of effect on short position is dependent on a flag to open a short and if we are not alredy short cash is negative since you are purchasing the adjusted independet stock
    backtest.entry_short_cash[(backtest.short_entry==1) & ((backtest.positions.shift(1)==0) | (backtest.positions.shift(1)==1))]=-backtest.indep_price.shift(-1)
    #Cash effect on closing a longposition is dependet on the flag to close the long and if you are indeed long, cash is positive effect since you are selling the dependent stock + the short adjustement for the last day close
    backtest.exit_long_cash[((backtest.long_exit==1) & (backtest.positions.shift(1)==1))]=(backtest.dep_price.shift(-1) - backtest.indep_short_adj)
    #cash effect on closing the short positionis dependent on the flag and if you are short indded, it is positive effect on the cash since you are selling the adjusted independet stock + the short adjustment from the previous day
    backtest.exit_short_cash[((backtest.short_exit==1) & (backtest.positions.shift(1)==-1))]=(backtest.indep_price.shift(-1) - backtest.dep_short_adj)
    #Flag if a particular day has short adjustment or not, when you open a position there is no short adjustemnton the day
    backtest.short_nav[(backtest.entry_long_cash ==0.0) & (backtest.entry_short_cash==0)] = True
    #adding transaction costs
    backtest.entry_long_cash = backtest.entry_long_cash*(1+trans_cost)
    backtest.entry_short_cash=backtest.entry_short_cash*(1+trans_cost)
    backtest.exit_long_cash= backtest.exit_long_cash*(1-trans_cost)
    backtest.exit_short_cash=backtest.exit_short_cash*(1-trans_cost)
    #selecting the corret shprt adjustment in case the day has it
    backtest.short_nav[backtest.short_nav == True] = backtest.positions
    backtest.short_nav = backtest.short_nav.shift(1)
    backtest.short_nav[backtest.short_nav == 1] = -backtest.indep_short_adj
    backtest.short_nav[backtest.short_nav == -1] = -backtest.dep_short_adj
    backtest.long_nav[backtest.positions.shift(1)==1] = backtest.dep_price
    backtest.long_nav[backtest.positions.shift(1)==-1] = backtest.indep_price
    #shift some columns to mke the next day only trading rule valid
    backtest[['entry_long_cash','entry_short_cash','exit_long_cash','exit_short_cash']] =backtest[['entry_long_cash','entry_short_cash','exit_long_cash','exit_short_cash']].shift(1)
    
    backtest.fillna(0, inplace=True)
    
    #calculating the final cash, tried to vectorize this part but so far i was not able 
    backtest['final_cash'] = 0.0
    backtest['initial_cash'] = 0.0
    for n in range(0,backtest.shape[0]):
        backtest.initial_cash[n] = initialcash
        finalcash = initialcash+backtest.entry_short_cash.iloc[n]+backtest.entry_long_cash.iloc[n]+backtest.exit_long_cash.iloc[n]+backtest.exit_short_cash.iloc[n]
        initialcash = finalcash
        backtest.final_cash.iloc[n] = finalcash
    backtest['nav']= backtest.final_cash + backtest.short_nav +  backtest.long_nav
    
    #creating a shorter version of the returning dataframe    
    backtest_short = backtest[['ITUB4','ITUB3','IBOV','nav']]
    
    if show_entire_df:
        return backtest
    elif not show_entire_df:
        return backtest_short

def quick_zspread(ts,lookback):
    
    #just a function when you want to calculate just the zspread
    
    ts['moving_average'] = ts.sprd.rolling(lookback).mean()
    ts['moving_std_dev'] = ts.sprd.rolling(lookback).std()
    ts=ts.dropna()
    ts['z_score'] = (ts.sprd- ts.moving_average)/ts.moving_std_dev
    
    return ts
    
    
# Function that defines the hedge ratio considering data for the last reb_lookback days

def optm_strategy(initial_series,reb_lookback,rebalance,adjhl,zentry,zexit):
    for n in range(cadf_lookback,initial_series.shape[0],rebalance):

        #Defining current execution range and next estimation range
        temp_executiondf =initial_series[n-adjhl+1:n+rebalance][['ITUB4','ITUB3','IBOV']]
        temp_estimationf =initial_series[n+rebalance-cadf_lookback:n+rebalance][['ITUB4','ITUB3','IBOV']]

        if n+cadf_lookback >= initial_series.shape[0] : break #stop the loop when future series is higher that end of dataseries

        #Executing strategy for execution current range    
        temp_executiondf['sprd'] =temp_executiondf.ITUB4-temp_executiondf.ITUB3*hedge_ratio
        strategy_temp = gen_strategy(temp_executiondf,adjhl,zentry,zexit)
        if n == cadf_lookback: 
            finalseries = strategy_temp
        else :
            strategy_temp.iloc[0][['positions_short','positions_long','positions']]=finalseries.iloc[-1][['positions_short','positions_long','positions']]
            finalseries = pd.concat([finalseries,strategy_temp],axis=0)

        #Estimating new ZEntry and ZExit
        itub4_range_temp = np.array(temp_estimationf.ITUB4.tolist())
        itub3_range_temp = np.array(temp_estimationf.ITUB3.tolist())
        reg_results_temp = pair_regression(itub4_range_temp,itub3_range_temp,'ITUB4','ITUB3',False,False,True)

        if reg_results_temp[0]<= -2.57 :
            zentry = z_entry_stationary
            zexit = z_exit_stationary
        else :
            zentry = z_entry_nonstationary
            zexit = z_exit_nonstationary
  
    #This is step is necessary to correctally conect rebalance periods, the objective is that the new rebalances carries the last position of the last day of the previous rebalance
    while len(finalseries.positions.iloc[1:][((finalseries.positions == 0) & (finalseries.positions.shift(1) != 0)) & ((finalseries.long_entry == 0) & (finalseries.short_entry == 0))])>0:
        finalseries.positions.iloc[1:][((finalseries.positions == 0) & (finalseries.positions.shift(1) != 0)) & ((finalseries.long_entry == 0) & (finalseries.short_entry == 0))]=finalseries.positions.shift(1)

    return finalseries
```

### Cointegration Testing for Stocks Pair (ITUB4/ITUB3)

In order to thes for stationarity of time series we implement the The Engle-Granger procedure, the implementation steps are described on the next section.

#### The Engle-Granger Implementation Steps:

**Step 1** - Estimate the regression: $y_t = \alpha + \beta x_t +e_t$ (1), where:

$y_t$ is the dependent variable time series of prices

$\alpha$ is the intercept of the regression

$\beta$ is the multiplier of the independent variable, refereed on this chapter as Hedge Ratio

$e_t$ is the residuals (or error term) of the regression

In order to determine which is the most appropriate relationship between the stocks is important to test both stocks as dependent and independent variables, which leads us to 2 regressions for the example:
$$ITUB4 = \alpha +\beta IUTB3 +e_t\,(2)$$ (2)
$$ITUB3 = \alpha + \beta ITUB4 +e_t\,(3)$$

We consider $\alpha=0 (the intercept)$ the statistical reason behind this is that the linear regression model fits 2 variables one is  hedge ration and the other the intercept, by setting the intercept to 0, the model will only fit the hedge ratio which reduces "overfitting".

**Step 2** - Calculate the residuals (error term) $e_t$. That is, given that we consider $\alpha=0$ in (1), we can write the residuals as $e_t = y_t -\beta x_t$ (4), for the example the residuals would become:

$$e_{t1} = ITUB4 - \beta ITUB3$$
$$e_{t1} = ITUB3 - \beta ITUB4$$ 

**Step 3** - Perform a unit root test on $e_t$ to determine if it is I(0). The augmented Dickey-Fuller (CADF) test for stationarity on residuals (4) provides the unit root test required, the hypothesis H0 is that **the prices series are not stationary**. The test has to less than the critical value for H0 to be rejected and to conclude that the timeseries is stationary and not a random walk.

#### ADF Test - Mathematical background

The general set up for a CADF Tets is : $$\Delta Z_t = \beta_0 + \beta_1t + \delta Z_{t-1} + \sum\limits_{i=1}^k \phi_i \Delta Z_{t-1} + \epsilon_t$$

$k$ represents the lag in the time series and BIC or AIC can be used to get the optimal lag k, for the example in this chapter the model with lag order 1 is used, additionally the ADF Statistics varies depending on whether an intercept term $(\beta)$ are included, as explant on the The Engle-Granger Implementation Steps the spread should have zero mean and we set the intercept term to 0, but it is noted that some practitioners tend to keep the $(\beta)$. For our implementation the ADF formulation used is: $$\Delta Z_t=\delta Z_{t-1} + \phi \Delta Z_{t-1} + \epsilon_t$$

The null hypothesis (H0) of the ADF test is that the spread $\delta Z_{t-1}$ is a unit-root process, and the alternative (H1) is that the process is a stationary process :

H0 : $\delta$ = 1

H1 : $\delta$ < 1

The ADF Statistic is defined as the T-Statistic for the $\delta$ coefficient. For this example the ready implementation of the test present on the **statsmodels.tsa** package in Python is used and the **critical values** for the test statistic at the 1 %, 5 %, and 10 % levels are based on Based on MacKinnon (2010) :

*   1pct = 3.51%
*   5pct = 2.89%
*   10pct = 2.58%

### Implementation of the Strategy

#### Half life of mean reversion

In *Algorithmic Trading: Winning Strategies and Their Rationale (Wiley Trading), Ernie Chan* suggests a Bollinger band approach where trades are entered when price deviates by more than $x$ standard deviations from the mean, where $x$ is a parameter to be optimized. The lookback period for the rolling mean and standard deviation can be optimized or set to the half life of mean reversion.

Half life can be estimated from a Ornstein-Uhlenbeck stochastic process, which is a mean-reverting process,  described by the SDE :

$$dX_t = \alpha\, (\mu - X_t)\, dt + \sigma\, dW_t$$

where α > 0 and Wt is the Wiener process. It can easily be solved explicitly:

$$\displaystyle X_t = e^{-\alpha t} X_0 + \mu\, (1 - e^{-\alpha t}) + \int_0^t \sigma\, e^{\alpha(s-t)}\, dW_s.$$

So we deduce that

$$\displaystyle \begin{aligned} \mathsf EX_t &= e^{-\alpha t} X_0 + \mu\, (1 - e^{-\alpha t}) = \mu + o(1), \quad \mbox{ as } t\to\infty, \\ \mathrm{Var}\, X_t &= \frac {\sigma^2}{2\alpha} \left( 1 - e^{-2\alpha t} \right) = \frac {\sigma^2}{2\alpha}, \quad \mbox{ as } t\to\infty. \end{aligned}$$

The coefficient α is called the speed of mean reversion.

**Half-life of the mean-reversion, t1/2, is the average time it will take the process to get pulled half-way back to the mean**. To this end, we consider the ODE $\dot x = \alpha\, (\mu - x)$, which has the solution x(t) = μ + e-αt (x0 – μ). So we can find the half-time from the equation

$\displaystyle x(t_{1/2}) - \mu = \frac {x_0 - \mu}2,$

i.e. $\displaystyle t_{1/2} = \frac {\log 2}\alpha$. In particular, **the higher the mean-reversion speed is, the smaller is the half-life.**

Calculating the half-life for the spread we got : 

```python
portfolio_df = pd.DataFrame(portfolio)
portfolio_df.columns=['sprd']
portfolio_df['ITUB4'] = initial_series[['ITUB4']][252:]
portfolio_df['ITUB3'] = initial_series[['ITUB3']][252:]
portfolio_df['IBOV'] = initial_series[['IBOV']][252:] 
hl = int(half_life_cadf(portfolio_df))
print('Half-Life for the spread-portfolio is :',hl,'Days')
```

### Trading Strategy

To implement an example trade strategy we are going to use a technique known as Bollinger Bands$^*$. It involves taking a rolling simple moving average (SMA) of a price series and then forming "bands" surrounding the series that are a scalar multiple of the rolling standard deviation of the price series. The lookback period for the moving average and standard deviation is identical and we are setting it to the calculated half life above.

The trades will be generated based on calculated z-score (also known as a standard score) of the current latest spread price. This is achieved by taking the latest portfolio market price, subtracting the rolling mean and dividing by the rolling standard deviation. Once this z-score is calculated a position will be opened or closed out under the following conditions:

    • zscore < −zentry: Long entry
    • zscore > +zentry: Short entry
    • zscore >= −zexit: Long close
    • zscore <= +zexit: Short close

Where zscore is the latest standardised spread price, zentry is the trade entry threshold and zexit is the trade exit threshold.

Note that if zexit = 0, this means that a trade is closed when the price mean-reverts to the current mean (calculated as a rolling mean with lookback equal the halflife). If zexit = −zentry, the position is closed when the prices moves beyond the opposite band so as to as to trigger a trading signal of the opposite sign.

Since we are attributing the lookback period of the analysis to the **half life** of the spread, to finish implementing the strategy we need to define **zentry** and **zexit**. For the initial exmaple we are going to visually define it looking at the ZSpread chart for the first 252 days. It is possible to see that the Z_Spread varies most of the time betwwen -2 and 2, so we are setting *zentry=2* and *zexit = -2*, so the close signal matchs the enter signal, meaning we are always positioned ont he spread.

$^*$ *Implementation described on Algorithmic Trading: Winning Strategies and Their Rationale (Wiley Trading) by Ernie Chan, and Advanced Algorithmic Trading by Michael L. Halls Moore*

### Back Test of the Trading Strategy

Back test the strategy was a real challenge and it was hard to settle on proposed approachs that were found during research, some of the approachs proposed overlly simplistic assumptions that produced really inexplicable results, and others highlly complex that required a lot of data that it is not avaiable easially, especially on a Emerging Market market.
The approach proposed is a NAV (Net Asset Value) based one with some simplifiying assumptions but which i believe holds similarities to real execution :


<dl>
<dd>1. The trades are executed on the Close Price of the Assets</dd>
<dd>2. But, when a entry/exit point is flagged on the strategy, you can only execute the trade on the next close price, here the rational is that you used today's close price as an input on your strategy to get the entry/exit flag, so you can only execute the flag on the next trading day </dd>
<dd>3. The inital cash of the strategy is set to be equal to the unit value of the first trade of the strategy</dd>
<dd>4. There is no penalty for holding a "negative cash balance" but it discounts from the NAV Value</dd>
</dl>

Following those rules the daily NAV calculation is as follows :

<dl>
<dd>1. Long Position : NAV is the value of your stock holdings </dd>
<dd>2. Short Position : NAV is zero since the cash proceeds from the sale balances the liabilities of the short holdings, but there is a daily adjustment to NAV based on asset/liability of the short position</dd>
<dd>3. Final NAV value daily : Final Cash of the Day + Long Position Value + Asset/Liability of the Short Position</dd>
<dd>4. Daily Strategy Return is $(NAV/NAV_{(t-1)})-1$ </dd>
</dl>

The interesting part of the approach is that given the Strategy Entry/Exit points defined on the previous step, the NAV calculation is almost entirelly vectorized which provides a very efficient calculation with transaction costs easyaly incorporated at each trade. ### Optimizing the Strategy

#### Look Ahead Bias

One important point to avoid on the set up of the strategy is the **_Look Ahead Bias_** which is the bias created by the use of information or data in a study or simulation that would not have been known or available during the period being analyzed. This will usually lead to inaccurate results in the study or simulation. Look ahead bias was introduced on the last strategy on purpose, the half life calculation utilised the information from the entire dataset, which would not be avaiable at the time of implemenattion.

The objective of the following sections is to propose an optimised strategy that avoids any **look ahead bias".

```python
rebalance = 30
cadf_lookback = 90
z_entry_stationary = 1
z_exit_stationary = -1
z_entry_nonstationary = 2
z_exit_nonstationary = -2
```

The initial adjusments accoridng to the first 90 days of data are :

```python
series=initial_series[:cadf_lookback][['ITUB4','ITUB3','IBOV']]
itub4_range_temp = np.array(series.ITUB4.tolist())
itub3_range_temp = np.array(series.ITUB3.tolist())
reg_results_temp = pair_regression(itub4_range_temp,itub3_range_temp,'ITUB4','ITUB3',False,False,True)
hedge_ratio = reg_results_temp[3]
print("CADF T-Stat = ",reg_results_temp[0], "Suggesting Cointegration for the selected Range")
print("Selected Portfolio is = ",reg_results_temp[2])

temp_series=initial_series[:cadf_lookback][['ITUB4','ITUB3','IBOV']]
temp_series['sprd'] = temp_series.ITUB4 - temp_series.ITUB3*reg_results_temp[3]

hl = int(half_life_cadf(temp_series))
if hl < 10 : adjhl = 20
elif hl >100: adjhl = 100
print('Calculated Half Life is : ',hl,' Adjusted Half Life is :',adjhl)

if reg_results_temp[0]<= -2.57 :
    zentry = z_entry_stationary
    zexit = z_exit_stationary
else :
    zentry = z_entry_nonstationary
    zexit = z_exit_nonstationary
    
print('Initial Z_Entry : ',zentry,'Initial Z_Exit :  :',zexit)
```

#### Final Conclusions

In this chapter we simulated a simple long/short strategy which is based on a cointegrated series, the series itself is the product of a linear combination between 2 assets. Different share classes from the same company are suitable for this type of analysis, especially in emerging markets which this type of situation tends to take more time to be adjusted, in our example we simulated 10 years of the strategy with very reasonable results. Of course, in practice a lot of the assumptions don't hold and the performance can be very different from the one predicted by the model, but still it is a relatively simple strategy to be implemented which i think compensates the time spent on researching this kind of opportunities.
A further development of this strategy would be the addition of a third asset which we believe holds a relationship with the basic strategy, for our example it is accepted by the market that the difference betwwen the ON and PN share classes in Brazil are due to liquidity differences in the assets (as explanied on the introduction) a natural step would be the introduction of a asset (preference for a tradable one) that could be used as a proxy of market liquidity.

For the pair of shares used in the anaöysis we used the CADF framework for stationarity test, one of the biggest drawbacks of the test was that it was only capable of being applied to two separate time series. With the addition of a third asset we need another procedure for stationarity test, for it we would use Johansen that allows us to determine if three or more time series are cointegrated. We will then form a stationary series by taking a linear combination of the underlying series and them create the strategy.
